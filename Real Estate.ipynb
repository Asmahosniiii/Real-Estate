# Data Visualzation

```python
df.hist(color='#555870',figsize=(17,14))
plt.show()
```

```python
df['BuildingArea'].value_counts().head(5).plot.bar(color='#555870')
```

```python
df['YearBuilt'].value_counts().head(5).plot.bar(color='#555870')
```

```python
plt.figure(figsize=(4,4))
colors = ['#E6E6E6', '#555870', '#C0BABA', '#7E7B8D','#DE6A73']
labels= ['Sold','Sold Prior','Passed In', 'Vendor Bid','Sold After Auction']
plt.pie(df['Method'].value_counts(), labels=labels,colors=colors,  autopct='%0.f%%', pctdistance=1.15, labeldistance=1.4)
plt.title('Sum of Methods')
plt.show()
```

```python
plt.figure(figsize=(4,4))
colors = ['#E6E6E6', '#555870', '#C0BABA', '#7E7B8D','#DE6A73']
labels= ['House','U','Townhouse']
plt.pie(df['Type'].value_counts(),labels=labels, colors=colors,autopct='%1.1f%%', pctdistance=0.85)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Count of Types')
plt.show()
```

```python
sns.countplot(df, x='BuildingArea', hue= 'Type',color="#555870", order=df.BuildingArea.value_counts().iloc[:1].index)
plt.title('BuildingArea By Type')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.hist(df['YearBuilt'],histtype='bar',color="#555870",alpha=1, bins=50)
plt.hist(df['Price'],color="#555870",histtype='bar',alpha=1, bins=50)
plt.show()
```
